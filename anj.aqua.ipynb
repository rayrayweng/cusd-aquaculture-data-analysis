```python
import pandas
data=pandas.read_csv('DesignSprintData.csv')
data=data[data.water_temp!=-127.0]
mean= data["water_temp"].mean()
std= data["water_temp"].std()
print(mean)
print(std)
```

```python
accum=0
values = []
for x in data["water_temp"]:
    if (abs(mean-x) > 4*std):
        accum+=1
        values.append(x)
print(accum)
print(values)
```
